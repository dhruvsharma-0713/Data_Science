## Setup and Data
Let's start with the data for the assignment. A pie chart is used to show parts of a whole, so we'll use percentages that add up to 100.

- os_market_share = [65, 25, 10]
- labels = ['Android', 'iOS', 'Others']

## Level 1: Pie Chart
Goal: Create a pie chart showing the market share of mobile OS.

The Concept: A pie chart is created using the plt.pie() function. Key parameters are:

- `labels`: To label each slice.
- `autopct`: To display the percentage on each slice.
- `explode`: To pull one or more slices out from the center for emphasis.

```python
import matplotlib.pyplot as plt
```

```python
os_market_share = [65, 25, 10]
labels = ['Android', 'iOS', 'Others']
explode = (0.1, 0, 0)
```

```python
plt.pie(os_market_share, 
        labels=labels, 
        autopct='%1.1f%%',
        explode=explode,
        shadow=True,
        startangle=90)
plt.title("Mobile OS Market Share")
plt.axis('equal')
```

## Level 2: Horizontal Bar Chart Comparison
Goal: Recreate the same data as a horizontal bar chart and observe which is easier to understand.

```python
plt.barh(labels, os_market_share, color=['limegreen', 'silver', 'grey'])
plt.xlabel("Market Share (%)")
plt.ylabel("Operating System")
plt.title("Mobile OS Market Share")
```

#### Observation: Which chart was better?

- The pie chart is great for quickly seeing the overall distribution and which category is the largest part of the whole.

- The bar chart is often better for making precise comparisons. It's much easier to see that Android's market share is more than double that of iOS. For many data analysts, bar charts are preferred over pie charts for this reason.
