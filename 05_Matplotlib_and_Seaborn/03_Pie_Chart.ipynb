```python
import matplotlib.pyplot as plt

plt.style.use("ggplot")
```

```python
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
```

```python
plt.title("Career Runs of Indian Batsmen")
plt.pie(runs, labels=labels)
```

```python
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.pie(runs, labels=labels, colors=colors)
```

```python
plt.pie(
    runs,
    labels=labels,
    colors=colors,
    wedgeprops={'edgecolor': 'black', 'linewidth': 2, 'linestyle': '--'}
)
```

```python
explode = [0.1, 0, 0, 0]

plt.pie(runs, labels=labels, explode=explode, colors=colors)
plt.title("Exploded Pie Example")
```

```python
plt.pie(
    runs,
    labels=labels,
    explode=explode,
    colors=colors,
    shadow=True
)
```

```python
plt.pie(
    runs, labels=labels, explode=explode, colors=colors, shadow=True, startangle=140
)
```

```python
plt.pie(
    runs, labels=labels, explode=explode, colors=colors, shadow=True, startangle=90, autopct='%1.1f%%'
)
```

```python
plt.pie(
    runs,
    labels=labels,
    autopct='%1.1f%%',
    startangle=140
)
plt.title("Career Run Share")
```
