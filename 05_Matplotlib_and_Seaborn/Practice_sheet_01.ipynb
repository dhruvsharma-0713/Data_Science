### Your Mission: Create a plot that helps you answer the following questions visually:

    - Did all three companies generally trend up or down during the week?

```python
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
company_A_price = [150, 152, 155, 151, 154]
company_B_price = [130, 135, 133, 138, 136]
company_C_price = [200, 201, 198, 203, 205]
```

```python
with plt.style.context('fivethirtyeight'):
    plt.plot(days, company_A_price, marker='o', label="Company A")
    plt.plot(days, company_B_price, marker='s', label="Company B")
    plt.plot(days, company_C_price, marker='^', label="Company C")

    plt.xlabel("Day of the Week")
    plt.ylabel("Stock Price ($)")
    plt.title("Weekly Stock Price Performance")
    plt.legend()
```
