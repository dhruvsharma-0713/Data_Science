## Level 1: Side-by-Side (Grouped) Bar Chart

Goal: Create a chart comparing the runs of Sachin, Kohli, and Sehwag for 5 selected years.

The Concept: To create a grouped bar chart, we need a little trick. We'll define the position for each year on the x-axis. Then, for each player, we'll plot their bar slightly to the left, right, or in the center of that year's position. This creates the "side-by-side" effect.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
tendulkar = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
```

```python
years_selected = years[:5]
tendulkar_selected = tendulkar[:5]
kohli_selected = kohli[:5]
sehwag_selected = sehwag[:5]
```

```python
x = np.arange(len(years_selected))
width = 0.25
```

```python
plt.bar(x - width, tendulkar_selected, width, label='Tendulkar')
plt.bar(x, kohli_selected, width, label='Kohli')
plt.bar(x + width, sehwag_selected, width, label='Sehwag')
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Run Comparison: First 5 Selected Years")
plt.xticks(x, years_selected)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
```

## Level 2: Horizontal Bar Chart

Goal: Show the total runs scored by each player in their debut 5 years.

The Concept: A horizontal bar chart (plt.barh()) is excellent for comparing totals across different categories, especially when the category names (the players' names) are long. It's much easier to read "Tendulkar", "Kohli", etc., on the y-axis than squished along the x-axis.

```python
players = ['Tendulkar', 'Kohli', 'Sehwag']
total_runs = [sum(tendulkar[:5]), sum(kohli[:5]), sum(sehwag[:5])]
```

```python
plt.barh(players, total_runs, color=['blue', 'red', 'green'])
plt.xlabel("Total Runs")
plt.ylabel("Player")
plt.title("Total Runs in Debut 5 Years")
```
