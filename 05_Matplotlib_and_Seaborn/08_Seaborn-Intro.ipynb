```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
sns.set_theme(style="darkgrid")
```

```python
x = np.array([0, 2, 71, 3, 78, 4, 5, 34, 6, 7, 1, 8, 43, 9 ,10, 89, 60])
y = np.sin(x)
```

```python
sns.lineplot(x=x, y=y)
plt.title('Beautiful Line Plot')
```
