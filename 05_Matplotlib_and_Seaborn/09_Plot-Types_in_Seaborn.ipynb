```python
import seaborn as sns
import matplotlib.pyplot as plt
print(sns.get_dataset_names())
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.info()
```

```python
sns.lineplot(x='total_bill', y='tip', data=tips)
plt.title('Line Plot')
plt.xlabel('Bill_Paid')
plt.ylabel('Tip (in `$`)')
plt.tight_layout()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Line Plot')
plt.xlabel('Bill_Paid')
plt.ylabel('Tip (in `$`)')
plt.tight_layout()
```

```python
sns.barplot(x='total_bill', y='tip', data=tips)
plt.title('Line Plot')
plt.xlabel('Bill_Paid')
plt.ylabel('Tip (in `$`)')
plt.tight_layout()
```

```python
sns.boxplot(x='total_bill', y='tip', data=tips)
plt.title('Line Plot')
plt.xlabel('Bill_Paid')
plt.ylabel('Tip (in `$`)')
plt.tight_layout()
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
pv= flights.pivot(index='month', columns='year', values='passengers')
pv
```

```python
sns.heatmap(pv, annot=True, fmt='d', cmap='magma')
```

```python
sns.heatmap(pv, annot=True, fmt='d', cmap='magma')
plt.title('passengers per month, Yearly')
plt.xlabel('Years')
plt.ylabel('Months')
```
