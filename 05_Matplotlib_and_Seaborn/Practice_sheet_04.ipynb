## Setup and Data
The Concept: Stack plots (or area charts) are perfect for visualizing how the composition of a group changes over a continuous variable, like time. We'll use `plt.stackplot()` to show how you might spend your 24 hours over a typical week.

First, let's create some data for a weekly schedule.

#Days of the week for the x-axis
- days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']

#Hours spent on each activity per day (each list must have 7 values)
#Note: The sum for each day should ideally be 24 hours.
- studying = [8, 9, 8, 10, 6, 2, 1]
- sleeping = [8, 8, 8, 7, 8, 10, 10]
- playing =  [2, 1, 2, 1, 4, 6, 7]
- other =    [6, 6, 6, 6, 6, 6, 6]

#Labels for the legend
- labels = ['Studying', 'Sleeping', 'Playing', 'Other']

## Task: Weekly Schedule Stack Plot
Goal: Create a stack plot of your weekly schedule.

```python
import matplotlib.pyplot as plt
```

```python
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
studying = [8, 9, 8, 10, 6, 2, 1]
sleeping = [8, 8, 8, 7, 8, 10, 10]
playing =  [2, 1, 2, 1, 4, 6, 7]
other =    [6, 6, 6, 6, 6, 6, 6]
labels = ['Studying', 'Sleeping', 'Playing', 'Other']
```

```python
plt.stackplot(days, studying, sleeping, playing, other, labels=labels)
plt.xlabel("Day of the Week")
plt.ylabel("Hours Spent")
plt.title("My Typical Weekly Schedule")
plt.legend(loc='upper left')
```
